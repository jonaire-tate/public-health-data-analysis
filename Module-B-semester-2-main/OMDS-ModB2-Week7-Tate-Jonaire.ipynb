# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

Several preattentive attributes could be used to direct the viewer’s attention to the last three months of data first. One approach would be to highlight the final three data points using a different color, such as red or orange, while keeping earlier data in a neutral color like gray. Another option would be to increase the thickness or size of the line or markers for the last three months so they stand out visually. A third approach could be to add annotations or labels to the last three data points to draw the viewer’s attention to them. Additionally, increasing contrast by fading earlier months and emphasizing the last three months would guide the viewer’s eye toward the most recent data.

```python
import matplotlib.pyplot as plt

# Create the dataset
data = {
    "A": [1,1,2,1,3,3,4,2,1,4,2,7,1,1],
    "B": [2,3,3,2,2,1,1,4,5,3,3,2,2,2],
    "C": [3,4,6,8,4,6,8,10,4,7,4,3,3,4],
    "D": [6,5,12,4,8,5,10,9,6,9,5,4,6,3],
    "E": [7,6,8,7,10,4,5,8,3,12,11,8,7,5]
}

countries = [
    "Australia","Brazil","Canada","China","France",
    "Germany","India","Italy","Mexico","Russia",
    "Spain","Turkey","United Kingdom","United States"
]

df = pd.DataFrame(data, index=countries)

# Plot heatmap
plt.figure(figsize=(8,6))

sns.heatmap(
    df,
    annot=True,
    cmap="Blues_r",   # darker color = higher rank emphasis
    cbar=False,
    linewidths=0.5
)

plt.title("Top 5 Drugs: Country-Level Sales Rank")
plt.xlabel("Drug")
plt.ylabel("Country")

plt.show()
```
