```python
labels = ['Bored', 'Not great', 'OK', 'Kind of interested', 'Excited']
sizes = [11, 5, 40, 25, 19] 

plt.figure(figsize=(6,6))
wedges, texts, autotexts = plt.pie(
    sizes, 
    autopct='%1.0f%%', 
    startangle=140
)

plt.legend(
    wedges,
    labels,
    title="Responses",
    loc="center left",
    bbox_to_anchor=(1, 0.5)
)

plt.title('Survey Results')
plt.show()
```
